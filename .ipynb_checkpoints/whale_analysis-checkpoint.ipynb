## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_csv = Path("whale_returns.csv")

# YOUR CODE HERE
whale_returns = pd.read_csv(whale_returns_csv, parse_dates=True, index_col = 'Date', infer_datetime_format=True)
whale_returns.head(5)
```

```python
# Count nulls
# YOUR CODE HERE
whale_returns.isnull().sum()
```

```python
# Drop nulls
# YOUR CODE HERE
whale_returns = whale_returns.dropna()
whale_returns.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("algo_returns.csv")

# YOUR CODE HERE
algo_returns = pd.read_csv(algo_returns_csv, parse_dates=True, index_col = 'Date', infer_datetime_format=True)
algo_returns.head(5)
```

```python
# Count nulls
# YOUR CODE HERE
algo_returns.isnull().sum()
```

```python
# Drop nulls
# YOUR CODE HERE
algo_returns = algo_returns.dropna()
algo_returns.isnull().sum()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("sp500_history.csv")

# YOUR CODE HERE
sp500_history = pd.read_csv(sp500_history_csv, parse_dates=True, index_col = 'Date', infer_datetime_format=True)
sp500_history.sort_index(inplace=True)
sp500_history.head(5)
```

```python
# Check Data Types
# YOUR CODE HERE
sp500_history.dtypes
```

```python
# Fix Data Types
# YOUR CODE HERE
# sp500_history.astype('float64')
sp500_history['Close'] = sp500_history['Close'].str.replace('$','').astype('float64')
sp500_history.head(5)
sp500_history.dtypes
```

```python
# Calculate Daily Returns
# YOUR CODE HERE
sp500_history = sp500_history.pct_change()
sp500_history.head(5)
```

```python
# Drop nulls
# YOUR CODE HERE
sp500_history = sp500_history.dropna()
sp500_history.head(5)
```

```python
# Rename Column
# YOUR CODE HERE
sp500_history.columns = ['S&P 500']
sp500_history.head(5)
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
# YOUR CODE HERE

combined_returns = pd.concat([whale_returns, algo_returns, sp500_history], axis = 'columns', join = 'inner')
combined_returns.tail(5)
```

## Performance

Calculate and Plot the daily returns and cumulative returns. Does any portfolio outperform the S&P 500?

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Plot the rolling standard deviation of the various portfolios along with the rolling standard deviation of the S&P 500 (consider a 21 day window). Does the risk increase for each of the portfolios at the same time risk increases in the S&P?
2. Construct a correlation table for the algorithmic, whale, and S&P 500 returns. Which returns most closely mimic the S&P?
3. Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. Does the portfolio seem sensitive to movements in the S&P 500?
4. An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the ewm with a 21 day half-life.

```python
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window
# YOUR CODE HERE

rolling_std_plot = combined_returns.rolling(window=21).std().plot(figsize=(25,10), title="21 Day Rolling Standard Deviation",
                             fontsize=15,xlim=(dt.date(2015,3,3),dt.date(2019,4,23)))
rolling_std_plot.xaxis.label.set_size(20);
rolling_std_plot.title.set_size(20);
rolling_std_plot.legend(loc=2, fontsize=15);
```

```python
# Construct a correlation table
# YOUR CODE HERE

corr_table = combined_returns.corr()
corr_table
sns.heatmap(corr_table, vmin=-1, vmax=1)
```

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
# (Your graph may differ, dependent upon which portfolio you are comparing)
# YOUR CODE HERE

# rolling_beta = rolling_cov / rolling_var
# combined_returns.columns
rolling_cov = combined_returns['BERKSHIRE HATHAWAY INC'].rolling(window=21).cov(combined_returns['S&P 500'])
rolling_var = combined_returns['S&P 500'].rolling(window=21).var()
rolling_beta = rolling_cov / rolling_var

beta_plot = rolling_beta.plot(figsize=(25,10), title="Berkshire Hathaway Inc. Beta", fontsize=15, 
                              xlim=(dt.date(2015,3,3),dt.date(2019,4,23)))
beta_plot.xaxis.label.set_size(20);
beta_plot.title.set_size(20);

rolling_beta.describe()
```

```python
# Calculate a rolling window using the exponentially weighted moving average. 
# YOUR CODE HERE

rolling_ewm = combined_returns.ewm(alpha=0.5, min_periods=21).mean()
rolling_ewm_plot = rolling_ewm.plot(figsize=(25,10), title="Exponentially Weighted Average",
                             fontsize=15,xlim=(dt.date(2015,3,3),dt.date(2019,4,23)))
rolling_ewm_plot.xaxis.label.set_size(20);
rolling_ewm_plot.title.set_size(20);
rolling_ewm_plot.legend(loc=3, fontsize=15);
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Calculate annualized Sharpe Ratios
# YOUR CODE HERE

sharpe_ratios = (combined_returns.mean() * 252 - 0) / (combined_returns.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
# YOUR CODE HERE

sharpe_ratios.plot(kind='bar', title='Sharpe Ratios')
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Read the first stock
# YOUR CODE HERE

aapl_csv = Path('AAPL.csv')
aapl = pd.read_csv(aapl_csv, parse_dates=True, index_col = 'Date', infer_datetime_format=True)
# x = aapl['Date'][0][:(len(aapl['Date'][0])-9)]
# datetime.strptime(aapl['Date'][0], '%m/%d/%y')
# aapl.index.dtype, combined_returns.index.dtype

aapl_returns = aapl.pct_change()
aapl_returns.head(5)
```

```python
# Read the second stock
# YOUR CODE HERE

tsla_csv = Path('TSLA.csv')
tsla = pd.read_csv(tsla_csv, parse_dates=True, index_col = 'Date', infer_datetime_format=True)
tsla_returns = tsla.pct_change()
tsla_returns.head(5)
```

```python
# Read the third stock
# YOUR CODE HERE

jnj_csv = Path('JNJ.csv')
jnj = pd.read_csv(jnj_csv, parse_dates=True, index_col = 'Date', infer_datetime_format=True)
jnj_returns = jnj.pct_change()
jnj_returns.head(5)
```

```python
# Concatenate all stocks into a single DataFrame
# YOUR CODE HERE

myport_returns = pd.concat([aapl_returns, tsla_returns, jnj_returns], axis = 'columns', join = 'inner')
myport_returns.columns = ['AAPL', 'TSLA', 'JNJ']
myport_returns.index = myport_returns.index.date
myport_returns.head(5)
```

```python
# Reset the index
# YOUR CODE HERE
```

```python
# Pivot so that each column of prices represents a unique symbol
# YOUR CODE HERE
myport_returns.pivot_table(index='AAPL', values=['TSLA', 'JNJ']).head(2)
```

```python
# Drop Nulls
# YOUR CODE HERE

myport_returns.isnull().sum()
myport_returns = myport_returns.dropna()
myport_returns.isnull().sum()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
# YOUR CODE HERE
weighted_returns = myport_returns.dot(weights)
weighted_returns.columns = ['Custom']
weighted_returns.head(5)
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns
# YOUR CODE HERE

added_combined_returns = pd.concat([combined_returns, weighted_returns], axis = 'columns', join = 'inner')
added_combined_returns.columns = [  'SOROS FUND MANAGEMENT LLC',          'PAULSON & CO.INC. ',
       'TIGER GLOBAL MANAGEMENT LLC',      'BERKSHIRE HATHAWAY INC',
                            'Algo 1',                      'Algo 2',
                           'S&P 500',                             'Custom']
added_combined_returns.tail(5)
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
# YOUR CODE HERE
added_combined_returns = added_combined_returns.dropna()
added_combined_returns.isnull().sum()
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others
